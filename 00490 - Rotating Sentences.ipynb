# Rotating Sentences

## 題目說明
在 **Rotating Sentences** 中，你需要將一系列輸入的句子 **順時針旋轉 90 度**。  
換句話說，原本由左至右、由上而下排列的文字，必須改為由上而下、由右至左排列。

## 輸入格式
- 輸入最多 **100 行**文字，每行最多 **100 個字元**。  
- 合法字元包含：換行、空白、任何標點符號、數字，以及大小寫英文字母。  
- 注意：**不允許 tab**。

## 輸出格式
- 將整批文字視為字元矩陣，輸出其 **順時針旋轉 90 度**的結果。  
- 規則：
  - **最後一行**會變成輸出時的**最左欄**。  
  - **第一行**會變成輸出時的**最右欄**。  
  - 若行長度不一，需在右側補空白至最長行長度再旋轉。  
  - 每行輸出後需**去除行尾多餘空白**。

## 範例

### Input

```
Rene Decartes once said,
"I think, therefore I am."
```

````python
from io import StringIO
import sys

sys.stdin = StringIO(
"""\
Rene Decartes once said,
"I think, therefore I am."
""")

lines = sys.stdin.read().splitlines()

maxL = 0
for i in range(len(lines)):
    maxL = max(maxL, len(lines[i]))   

for i in range(len(lines)):
    lines[i] = lines[i].ljust(maxL)

for j in range(maxL):
    row = ""
    for i in range(len(lines) - 1, -1, -1):
        row += lines[i][j]
    print(row.strip())
````
