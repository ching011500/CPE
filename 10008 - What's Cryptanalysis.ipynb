# Cryptanalysis

## 題目說明
Cryptanalysis 是破解加密文字的過程，本題要做的是對一段文字進行簡單的統計分析。  
給定一段輸入文字，請統計每個英文字母出現的次數，並依規則輸出結果。

## 輸入格式
- 第一行：一個正整數 `n`，代表接下來會有多少行文字。  
- 接下來 `n` 行：每行可能包含空白、標點或數字。  

規則：
- 只計算英文字母 **A–Z**。  
- 大小寫視為相同（例如 `a` 與 `A` 算同一個字母）。

## 輸出格式
- 每行輸出一個大寫字母與其出現次數，中間以空白隔開。  
- 排序規則：
  1. 出現次數由大到小。  
  2. 若次數相同，則依字母順序由小到大。  
- 沒有出現的字母不輸出。

## 範例

### Input

```
3
This is a test.
Count me 1 2 3 4 5.
Wow!!!! Is this question easy?
```

### Output

```
S 7
T 6
I 5
E 4
O 3
A 2
H 2
N 2
U 2
W 2
C 1
M 1
Q 1
Y 1
```

````python
from io import StringIO
import sys

sys.stdin = StringIO(
"""\
3
This is a test.
Count me 1 2 3 4 5.
Wow!!!! Is this question easy?
""")

N = int(sys.stdin.readline().strip())
count = [0] * 26
lines = sys.stdin.read()


for c in lines:
    if c.isalpha():
        c = c.upper()
        count[ord(c) - 65] += 1

L = []
for i in range(26):
    if count[i] > 0:
        L.append((chr(i + 65), count[i]))

L.sort(key = lambda x : (-x[1], x[0]))

for a, b in L:
    print(a, b)
````
